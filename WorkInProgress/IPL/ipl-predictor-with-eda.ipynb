```python
import seaborn as sns
import matplotlib.pyplot as plt
matches = pd.read_csv('/kaggle/input/ipl-complete-dataset-20082020/matches.csv')
deliveries = pd.read_csv('/kaggle/input/ipl-complete-dataset-20082020/deliveries.csv')
```

```python
totalrun_df = deliveries.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
totalrun_df
```

```python
totalrun_df = totalrun_df[totalrun_df['inning']==1]
totalrun_df['total_runs'] = totalrun_df['total_runs'].apply(lambda x:x+1)
totalrun_df
```

```python
matches_df = matches.merge(totalrun_df[['match_id','total_runs']],left_on='id',right_on='match_id')
matches_df
```

```python
matches_df['team1'].unique()
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
# replacing the Delhi Daredevils with Delhi Capitals

matches_df['team1'] = matches_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
matches_df['team2'] = matches_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')


# replacing the Deccan Chargers with Sunrises Hyderabad

matches_df['team1'] = matches_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
matches_df['team2'] = matches_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
matches_df = matches_df[matches_df['team1'].isin(teams)]
matches_df = matches_df[matches_df['team2'].isin(teams)]

matches_df['team1'].unique()
```
