```python
data['Total_Area'] = data['Area of the house(excluding basement)']+data['Area of the basement']
```

```python
data.drop(columns=['Lattitude','Longitude','Built Year','id','Date','Postal Code','number of views','Area of the house(excluding basement)','Area of the basement'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = data.drop(columns=['Price'])
y = data['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
model1 = LinearRegression()
model1.fit(X_train, y_train)
```

```python
Y_predict = model1.predict(X_test)
Y_predict
```

```python
print('MSE:', mean_squared_error(y_test, Y_predict))
```

```python

print('R2 Score: ',model1.score(X_test,y_test))
```

```python
model2 = RandomForestRegressor()
model2.fit(X_train, y_train)
```

```python
Y_predict = model2.predict(X_test)
Y_predict
```

```python
print('MSE:', mean_squared_error(y_test, Y_predict))
```

```python
print('R2 Score: ',model2.score(X_test,y_test))
```

```python
from sklearn.linear_model import RidgeCV
model3 = RidgeCV()
model3.fit(X_train, y_train)
```

```python
Y_predict = model3.predict(X_test)
```

```python
print('MSE:', mean_squared_error(y_test, Y_predict))
```

```python
print('R2 Score: ',model3.score(X_test,y_test))
```
